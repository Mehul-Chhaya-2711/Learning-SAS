POSITIONAL AND KEYWORD PARAMETERS

1. POSITIONAL PARAMETERS COME WITHOUT AN '=' SIGN LIKE %macro macro_name(var1,var2). Variables passed need to be in same order
2. KEYWORD PARAMETERS COME WITH AN '=' SIGN LIKE %macro macro_name(var1=,var2=). Variables passed need not be in same order
3. ALWAYS USE KEYWORD PARAMETERS

```sas
/*MACRO DEFINITION || IT RESTRICTS THE PRINTING OUTPUT AS PER THE VARIABLES PASSED IN MACRO*/
%macro print_var(var1=,var2=);
proc print data=sashelp.class;
var &var1 &var2;
run;
%mend print_var;


%print_var (var1=name,var2=weight);
```

MAKE REPITITIVE CODES A MACRO EXAMPLE

SIMPLE EXAMPLE

```sas
data class_lt_100;
set sashelp.class;
where weight lt 100;

data class_gt_100;
set sashelp.class;
where weight ge 100;

proc print data = class_lt_100;
proc print data = class_gt_100;
run;
```

MACRONISED VERSION

```sas
%macro class_dataset(weight_specification=,dataset_name=,weight_condition=);
data class_lt_&dataset_name;
set sashelp.class;
where weight &weight_condition &weight_specification;
proc print data = class_lt_&dataset_name;
run;
%mend class_dataset;

%class_dataset(weight_specification=100,dataset_name=100,weight_condition=lt);
%class_dataset(weight_specification=100,dataset_name=100,weight_condition=ge);
run;

/*THING TO REMEMBER|| SAS MACRO CAN BE REFERRED TO AS STRING REFERENCER THATS IT. IT REPLACES STRING IN THE CODES WHERE ONE WOULDNOT WANT TO DO 
IT EVERY NOW AND THEN TO AVOID MANUAL ERRORS. IT ALSO IS USED TO RUN CERTAIN CODE SNIPPETS REPETITIVELY*/
```

PROC SQL IN MACROS

```sas
proc means data = sashelp.heart noprint;
var height;
output out = out1 mean= mean_ht;
/*Here it is outputting entire proc means output to variable out1 and mean of heart to mean_ht*/
run;

proc sql noprint;
select mean_ht into : var1
from out1;
/*Here it is outputting meant of heart from out1 variable to varialble var1*/
quit;
%put &var1;
```

ITERATIVE MACROS AND CONDITIONAL MACROS

```sas
%macro iterate(NO);
    %do i=0 %to &NO;
        %if (&i=2 or &i=4 or &i=6 or &i=8 or &i=10) %then 
            %do; 
                %put even;
            %end;
        %else 
            %do;
                %put odd;
            %end;    
    %end;
%mend iterate;

%iterate(12);
```
