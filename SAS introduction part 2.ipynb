Creating Inline Data || Meaning ; creating data as we write the code

```sas
Data inline_data;                        /*The Data getting created here is a temporary data getting stored in WORK folder.
                                           It will vanish away upon session termination*/
    variable_1 = 2;
    variable_2 = 3;
proc print data = inline_data;    
run;    
```

Creating Inline Data Part 2 || Initializing the variables first and then punching in the data in seperate block of lines

```sas
Data inline_data_2;
    Input column_1 Column_2 char_column $10.;                /*There are only 2 data types: Char and Numeric; Char is initialized by variable_name 
                                                               followed by $(with numeric value of character length allowed if needed) 
                                                               and numeric are initialized as is*/
    Cards;                                                    /*Punching in the cards of sets of 2 numbers in a row*/
        1 2 char1
        3 4 char2
        5 6 char3
        7 8 char4
Proc print data = inline_data_2;    
Run;
```

Importing Data from a Raw File using Data Step

```sas
/*Creating a file name variable to avoid typing same file paths more than once*/
FILENAME file_1"/folders/myfolders/Practice_Folders_SAS/SFF_Dataset.csv";

data new_dataset;
    infile file_1
        dlm=','
        firstobs=2
        obs=5
        truncover
        ;
    length title $ 40;
    length author $ 40;
    length publication $ 40;
    length type $ 50;
    input title$ author$ publication$  type$ ;       /*A great learning here || Input statement (that specifies colun names) is a compulsory statement 
                                                         to be able to record some data and register some under a variable*/
proc print data = new_dataset;        
run;

/*Another important thing to learn is || Length of the variable needs to be specidifed before input statement in case of variable values being
greater than 8 characters in length. Because SAS stores only 8 characters by default and it needs to be given the length of variable in case
of the value bein > 8 characters*/
```

Importing Data from a Raw File using proc Statement

```sas
proc import 
     datafile="/folders/myfolders/Practice_Folders_SAS/SFF_Dataset.csv"
     dbms=dlm out=work.new_dataset     /*replaces above created dataset in temporary directory*/
     replace; 
     delimiter=','; 
proc print data = new_dataset;
run;
```

Difference between Format and Informat

informat is what we specify while writing the code such as $ for character variable, lengths etc.
format is what we specify for getting desired outputs

Following is an example

```sas
Data f_v_if;                        
    input name $ bday MMDDYY10. dday1 MMDDYY10. numbers;
    cards;
    mehul 11/27/199411/27/1994.
proc format;
value numbers .='NA';
proc print data = f_v_if;
    format bday MMDDYY10.;
run; 

/*Date variable is internally formatted to numeric type by substracting year 1960 from it; to reformat it to date type we need to apply format*/
```
