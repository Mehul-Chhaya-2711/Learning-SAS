MANAGING VARIABLES
1. Keep will keep a variable
2. Drop will drop a variable
3. firstobs will start at the mentioned position
4. obs will not return number of observations mentioned but will stop at that number of obdervation. Example:

1. IF applies the conditions after buffering the entire data into memory

3. WHERE applies in the data statement to subset
4. IF can be applied in INPUT statement

WHERE statement applies the subset condition before merging the data sets, Whereas, IF statement applies the subset condition after merging the data sets.

```sas
data ex1;
input ID Score;
cards;
1 25
2 28
3 35
4 45
;
proc print data = ex1;

data ex2;
input ID Score;
cards;
1 95
2 97
;
proc print data = ex2;

data comb1;
merge ex1 ex2;
by ID;
where score <= 30;
proc print data = comb1;

/*It returns 2 observations. WHERE condition applied before merging. It applies separately on each of the 2 data sets before merging.*/
data comb2;
merge ex1 ex2;
by ID;
if score <= 30;
proc print data = comb2;
run;

/*It returns 0 observation as IF condition applied after merging. Since there is no observation in which value of score is less than or equal to 30, it returns zero observation.
*/
```

Understanding the merge statement

```sas
DATA one;
   INPUT id v1 v2;
   DATALINES;
   1 10 100
   2 15 150
   3 20 200
   ;
  PROC SORT Data=one;
   BY id;
  
   DATA two;
   INPUT id v1 v2;
   DATALINES;
   1 1000 10000
   2 1500 15000
   3 2000 20000
   4  800 30000
   ;
  PROC SORT Data=two;
   BY id;
  
  DATA three;
   MERGE one two;
   BY id;
  PROC PRINT DATA=three; 
  RUN;
```

SIMPLE LOOPING USING DO END

```sas
DATA LOOPED_NUMBERS;
NUM = 0;
DO I = 1 TO 5;
    NUM = NUM + 1;
   OUTPUT;
END;
PROC PRINT DATA = LOOPED_NUMBERS;

DATA REVERSE_LOOPED_USING_BY;
DO I = 5 TO 1 BY -1;
    OUTPUT;
END;
PROC PRINT DATA = REVERSE_LOOPED_USING_BY;

DATA MONTH_NAMES_AND_NUMBERS;
MONTH_NUM = 0;
DO MONTH = 'JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEPT','OCT','NOV','DEC';
            MONTH_NUM = MONTH_NUM +1;
            OUTOUT;
END;
PROC PRINT DATA = MONTH_NAMES_AND_NUMBERS;

DATA DO_WHILE_DATASET;
ITERATOR = 5;
DO WHILE (ITERATOR > 0);
    ITERATOR = ITERATOR -1;
    OUTPUT;
END;
PROC PRINT DATA = DO_WHILE_DATASET;
RUN;

DATA DO_UNTIL_DATASET;
ITERATOR = 5;
DO UNTIL (ITERATOR > 0);
    ITERATOR = ITERATOR -1;
    OUTPUT;
END;
PROC PRINT DATA = DO_WHILE_DATASET;
RUN;
```

SAS ARRAYS

```sas
DATA TEST1;
INPUT A B C;
CARDS;
1 . 3
. 3 4
6 8 .
;
PROC PRINT DATA = TEST1;

/*Replaceing null with 0
This is better than If ELse statement as it eliminates the need to write IF else as many times as numbers of variables
/*
Steps:
1. Define array with number of elements; define elements of the array i.e. the column names
2. apply a DO loop
3. Write the statements that has necessary operations to be performed
4. end the DO loop
*/
DATA TEST_NO_NULL (drop=i);  /*This step will drop an additional variable I which gets created in the process*/
SET TEST1;
ARRAY X{3} A B C;
DO I = 1 TO 3;
IF X{I} = . THEN X{I} = 0;
END;
PROC PRINT DATA = TEST_NO_NULL;

/*Making the above statement more generic when we are not aware of the number of variables in a dataset*/
DATA TEST_NO_NULL_GEN (drop=i);  /*This step will drop an additional variable I which gets created in the process*/
SET TEST1;
ARRAY X{*} A B C;
DO I = 1 TO dim(X);  /*dim returns the number of elements in an array*/
IF X{I} = . THEN X{I} = 0;
END;
PROC PRINT DATA = TEST_NO_NULL_GEN;

/*USING 2 VARIABLES IN ARRAYS*/


/*USING 2 ARRAYS TO PICK EVERY SINGLE ELEMENTS OF MORE THAN ONE COLUMN AT A TIME AND CREATE A NEW VARIABLE*/
DATA TEST_DATA_TWO_VARIABLES;
INPUT A1 A2 A3 B1 B2 B3;
CARDS;
12 . 34 12 65 23
. 34 23 7 . 19
2 2 . 122 . .
;
PROC PRINT DATA = TEST_DATA_TWO_VARIABLES;

DATA TEST_DATA_TWO_VARIABLES_FINAL;
SET TEST_DATA_TWO_VARIABLES;
ARRAY A{3} A1 - A3;
ARRAY B{3} B1 - B3;
ARRAY C{3} C1 - C3;
DO I = 1 TO 3;
IF A{I} = . THEN A{i} = 0;
IF B{I} = . THEN B{i} = 999;
C{I} = SUM(A{I},B{I});
END;
PROC PRINT DATA = TEST_DATA_TWO_VARIABLES_FINAL;

/*LEAVE AND STOP STATEMENTS*/
DATA TEST_DATA_LEAVE;
SET TEST_DATA_TWO_VARIABLES;
ARRAY A{3} A1 - A3;
ARRAY B{3} B1 - B3;
ARRAY C{3} C1 - C3;
DO I = 1 TO 3;
IF A{I} = . THEN LEAVE ;
IF B{I} = . THEN B{i} = 999;
C{I} = SUM(A{I},B{I});
END;
PROC PRINT DATA = TEST_DATA_LEAVE;

DATA TEST_DATA_STOP;
SET TEST_DATA_TWO_VARIABLES;
ARRAY A{3} A1 - A3;
ARRAY B{3} B1 - B3;
ARRAY C{3} C1 - C3;
DO I = 1 TO 3;
IF A{I} = . THEN A{i} = 0;
IF B{I} = . THEN STOP;
C{I} = SUM(A{I},B{I});
END;
PROC PRINT DATA = TEST_DATA_STOP;

/*PLEASE NOTE: WE CANNOT WORK ON BOTH NUMERIC AND CHARACTER VARIABLES IN SAME ARRAT; BELOW IS THE EXAMPLE FOR CHAT ARRAY*/
DATA TEST1_CHAR;
INPUT A$ B$ C$;
CARDS;
1 . 3
. 3 4
6 8 .
;
PROC PRINT DATA = TEST1_CHAR;

DATA TEST_NO_NULL_CHAR (drop=i);  /*This step will drop an additional variable I which gets created in the process*/
SET TEST1_CHAR;
ARRAY X{3}$ A B C;
DO I = 1 TO 3;
IF X{I} = . THEN X{I} = "XX";
END;
PROC PRINT DATA = TEST_NO_NULL_CHAR;

/*USING KEYWORDS TO REFER TO ONLY NUMERIC OR ONLY CHARACTER VARIABLES*/
DATA TEST1_MIX;
INPUT A$ B$ C$ D E F;
CARDS;
1 . 3 1 . 3
. 3 4 . 3 4
6 8 . 6 8 .
;
PROC PRINT DATA = TEST1_MIX;

DATA TEST_NO_NULL_MIX (drop=i);  /*This step will drop an additional variable I which gets created in the process*/
SET TEST1_MIX;
ARRAY X{*} _NUMERIC_ ;
ARRAY Y{*} _CHARACTER_ ;
DO I = 1 TO DIM(X);
IF X{I} = . THEN X{I} = 0;
IF Y{I} = . THEN Y{I} = "XX";
END;
PROC PRINT DATA = TEST_NO_NULL_MIX;
RUN;
```

SORTING || TRANSPOSING THE DATA

```sas
DATA SALES_DATASET;
INPUT YEAR QUARTER SALES;
CARDS;
2019 1 123
2019 2 234
2019 3 456
2019 4 567
2020 1 432
2020 4 678
2020 2 999
2020 3 234
2020 3 234
2020 3 234
;
PROC PRINT DATA = SALES_DATASET;

/*PROC SORT IS A COMPULSORY STATEMENT BEFORE MERGING OPERATIONS (JUST TO REMEMBER)*/
/*OUT IS A COMPULSORY ARGUMENT ELSE IT WILL MAKE CHANGES IN ORIGINAL DATASET*/

PROC SORT 
    DATA = SALES_DATASET 
        OUT = SALES_DATASET_1 
              NODUPRECS; 
/*NODUPRECS WILL REMOVE DUPLICATE RECORDS*/
        BY DESCENDING SALES;
PROC PRINT DATA = SALES_DATASET_1;

PROC SORT 
     DATA = SALES_DATASET 
     OUT = SALES_DATASET_2 
     NODUPKEY;
/*NODUPKEYS WILL REMOVE RECORDS BASED ON THE VALUE OF KEY THAT IS GETTING REPEATED*/
     BY YEAR SALES;
PROC PRINT DATA = SALES_DATASET_2;

PROC TRANSPOSE DATA = SALES_DATASET_2 OUT = SALES_DATASET_TRANSPOSED(DROP=_NAME_) PREFIX=SALES_Q;
ID QUARTER;
VAR SALES;
BY YEAR;
PROC PRINT DATA = SALES_DATASET_TRANSPOSED;
RUN;
```

PROC SQL

1. USE CALCULATED TO REUSE THE NEW VARIABLES CREATED ON THE FLY
2. PLEASE KEEP IN MIND THAT SQL RUNS OVER SAS IN PROC SQL HENCE IT WILL SLOWER COMPARED TO SAS ITSELF

```sas
PROC SQL;
CREATE TABLE JUST_FOR_FUN AS 
SELECT YEAR FORMAT=8.2,
       QUARTER,SALES,
       CASE WHEN QUARTER = 1 THEN SALES*1.2
            WHEN QUARTER = 2 THEN SALES*1.3
            WHEN QUARTER = 3 THEN SALES*1.5
            ELSE SALES*1.4
            END AS AMOUNT_AFTER_PREMIUM,
       CALCULATED AMOUNT_AFTER_PREMIUM-SALES AS PROFIT
FROM SALES_DATASET
WHERE CALCULATED PROFIT > 100;

QUIT;
PROC PRINT DATA = JUST_FOR_FUN;
RUN;
```
